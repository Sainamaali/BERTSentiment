```python
# tokenizer is going to tansform a string to an array of numbers
# AutoModelForSequenceClassification from tarnsformesrs to load in our NLP model
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
#grab data from Yelp
import requests
from bs4 import BeautifulSoup
# create a function to extract specific comments we want
import re
```

# 2. Instantiate Model

```python
#creating tokenizer
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

# 3. Encode and Calculate Sentiment

Passing a string to tokenizer to test

```python
tokens = tokenizer.encode('It was good but couldve been better. Great', return_tensors='pt')
```

```python
tokens
```

```python
#re convert our tokenizer to the original sentence:
tokenizer.decode(tokens[0])
```

Passing the token to the model

```python
result = model(tokens)
```

Understanding the results: 
The SequenceClassifierOutput is a class from the Hugging Face Transformers library, used to hold the outputs of a sequence classification model. The fields in this output contain various types of information returned by the model.

loss: This field is None in your output, indicating that the loss was not computed. The loss is typically calculated during training when the labels are provided.

logits: The logits are the raw, unnormalized scores ** output ** by the last layer of the model before applying an activation function like softmax. In your example, the logits tensor is:

```python
result
```

This tensor represents the scores for each class for a single input sequence. Here, there are 5 classes, and the model has produced a score for each one. These logits can be converted to probabilities by applying a softmax function.
hidden_states: This field is None, which means that the intermediate hidden states of the model were not returned. Hidden states can be useful for inspecting the internal workings of the model or for tasks like probing.

attentions: This field is None, indicating that the attention weights were not returned. Attention weights can be useful for understanding which parts of the input the model focused on during processing, particularly in models like Transformers.


The shape of this tensor is [1, 5], meaning there is one sequence and five class scores.
Each value in the tensor represents the score for a corresponding class.
The grad_fn=<AddmmBackward0> part indicates that this tensor was generated as a result of a backward operation in the computational graph, which is used for automatic differentiation in PyTorch.

```python
#looking at logits which are the scores for each class (1-5 classes)
result.logits
```

getting the sentiment score between 1 to 5, the higher the more positive is the sentiment

```python
int(torch.argmax(result.logits))+1
```

# 4. Collect Reviews

```python
r = requests.get('https://www.yelp.com/biz/social-brew-cafe-pyrmont')
soup = BeautifulSoup(r.text, 'html.parser')
# extracting specific components that we want from this text page- it will grab classes that starts with comments
regex = re.compile('.*comment.*')
# find all the tags with the soup that matches our specific formatting- in a paragraph tag
results = soup.find_all('p', {'class':regex})
# Extarcting the text from the tags
reviews = [result.text for result in results]
```

```python
reviews
```

# 5. Load Reviews into DataFrame and Score

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.array(reviews), columns=['review'])
```

```python
df['review'].iloc[0]
```

```python
df
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

checking the sentiment score of the first sentence

```python
sentiment_score(df['review'].iloc[1])
```

running for all of the data in the dataframe

```python
df['sentiment'] = df['review'].apply(lambda x: sentiment_score(x[:512]))
```
